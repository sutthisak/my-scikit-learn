```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
print('OpenCV Version : %s ' % cv2.__version__)
```

```python
PATH = 'data/lena.png'
image = cv2.imread(PATH)
print(type(image))
print(image.shape)
plt.imshow(image)
plt.show()
```

The image has been correctly loaded by openCV as a numpy array, but the color of each pixel has been sorted as BGR.
Matplotlib's plot expects an RGB image

```python
image2 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image2)
plt.show()
```

```python
image3 = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image3 = cv2.cvtColor(image3, cv2.COLOR_GRAY2RGB)
plt.imshow(image3)
plt.show()
```
